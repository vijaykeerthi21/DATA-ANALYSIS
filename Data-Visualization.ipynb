```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#color's array
c = ['blue', 'green', 'red', 'yellow']
```

```python
#to read Csv file
df =pd.read_csv("auction[1].csv")
```

```python
#To check columns names
df.columns
```

```python
#Bar ploting for Type of player with count in df(Data Frame)
ax = sns.countplot(x ='TYPE',data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

# From above graph we can observe that
Majority players are bowlers and All rounders

batters and wicket keepers are lesser

```python
#TYPE of player with price paid
df.groupby(['TYPE'],as_index= False)['PRICE PAID'].sum().sort_values(by='PRICE PAID',ascending = False)
```

```python
# arraay of those price
x = [788500000,442000000,940500000,133500000]
```

```python
#Bar ploting for Type of player with PRICE PAID  in df(Data Frame)
players_role = df.groupby(['TYPE'],as_index = False)['PRICE PAID'].sum().sort_values(by = 'PRICE PAID', ascending= False)

sns.barplot(x = 'TYPE', y = 'PRICE PAID',data = players_role)

```

## From above graph we can see that most of the bowlers are highly paid then other type of players

Bowlers and All rounders are paid highly

Batters and Wicket keepers are paid low

```python
df.columns
```

```python
#Bar ploting for TEAMS with PRICE PAID in df(Data Frame)
players_price = df.groupby(['TEAM'],as_index = False)['PRICE PAID'].sum().sort_values(by = 'PRICE PAID', ascending= False)



sns.barplot(x = 'TEAM', y = 'PRICE PAID',data = players_price)
sns.set(rc = {'figure.figsize':(30,5)})
```

# This is the graphical represntation of team which are highly paid for players

```python
#Bar ploting for NATIONALITY  of player and TEAMS with count in df(Data Frame)
sns.countplot(data = df,x = 'TEAM', hue = 'NATIONALITY')


sns.set(rc = {'figure.figsize':(30, 5)})
```

# The above graph show's that players NATIONALITY  of players Team wise

```python
#Bar ploting for NAtionality of player with count in df(Data Frame)
ax = sns.countplot(data = df, x ='NATIONALITY')
sns.set(rc = {'figure.figsize':(5,5)})
```

```python
#price paid to nationality of players
grouped_by =df.groupby('NATIONALITY')['PRICE PAID'].sum()
print(grouped_by)
```

```python
# price paid to each team
grouped_by2 =df.groupby('TEAM')['PRICE PAID'].sum()
print(grouped_by2)
```

```python
#PIE CHART for NATIONALITY of players in df(Data Frame)
plt.pie(grouped_by.values,labels =grouped_by.index ,autopct="%.2f")
plt.show()
```

```python
#PIE CHART for TEAMS of players in df(Data Frame)
plt.pie(grouped_by2.values,labels =grouped_by2.index ,autopct="%.2f")
plt.show()
```

```python
#ploting of teams
plt.plot(df['TEAM'],df['PRICE PAID'], marker ="^" ,ls  ="--",color = "red",label =" TEAm")
plt.legend()
plt.show()
```
